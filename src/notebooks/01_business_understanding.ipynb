## 1️⃣ SAP P2P PROCESS OVERVIEW

### Process Flow:
```
PO (Purchase Order)
  ↓
GR (Goods Receipt - Livraison)
  ↓
IR (Invoice Receipt - Facture)
  ↓
GR/IR Clearing (Reconciliation)
  ↓
Payment
```

### Key Data Points:
- **PO:** Order from Supplier (amount, quantity, dates)
- **GR:** Goods physically received (po_history_category = 'E')
- **IR:** Invoice received from supplier (po_history_category = 'Q')
- **Matching:** GR must match IR for payment processing

### SAP Columns:
- `purchasing_document_|_ebeln` - PO Number
- `item_|_ebelp` - PO Line Item
- `po_history_category_|_bewtp` - Category (E=GR, Q=IR)
- `amount_|_wrbtr` - GR Amount
- `invoice_value_|_reewr` - IR Amount
- `supplier_|_lifnr` - Supplier ID

## 2️⃣ BUSINESS PROBLEM

### Current Challenges:
❌ **Unmatched GR/IR:** Invoices without goods receipts (fraud risk)  
❌ **Outstanding Goods:** Goods received but not invoiced (accounting risk)  
❌ **Amount Discrepancies:** GR amount ≠ IR amount (needs investigation)  
❌ **Manual Reconciliation:** Time-consuming, error-prone process  
❌ **Late Payment Holds:** Missing invoices delay payment cycles  

### Impact:
💰 **Financial:** Incorrect amount matching → payment delays/overages  
⚠️  **Operational:** Manual intervention required for exceptions  
🔴 **Compliance:** Unmatched transactions create audit findings  
🚨 **Fraud Risk:** IR without GR is major red flag

## 4️⃣ ANOMALY TYPES & SEVERITY

### Normal Flow:
| Type | Condition | Risk | Severity | Action |
|------|-----------|------|----------|--------|
| ✅ **OK** | GR ✓ + IR ✓ | None | NONE | Process normally |
| ⚠️ **DELIVERED_NOT_INVOICED** | GR ✓ + IR ✗ | Accounting | HIGH | Investigate delay |
| 🚨 **INVOICED_NOT_DELIVERED** | GR ✗ + IR ✓ | **FRAUD** | **CRITICAL** | **Block payment** |
| ❌ **INCOMPLETE** | GR ✗ + IR ✗ | Data quality | MEDIUM | Check PO status |

### Risk Thresholds:
- **Amount Gap >5%:** Flag for review
- **Outstanding >30 days:** Escalate
- **Supplier anomaly rate >20%:** High risk supplier
- **IR before GR:** Always suspicious
